```python
from newspaper import Article
```

```python
#Guardamos todo lo de la pagina en una variable
pagina = Article('https://es.wikipedia.org/wiki/Crist%C3%B3bal_Col%C3%B3n')
pagina.download()
pagina.parse()
```

```python
pagina.text
```

```python

#Importamos librerías
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS
```

```python
imagen = WordCloud()
imagen.generate(pagina.text)
plt.imshow(imagen, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
#Importamos librería
from PIL import Image
import numpy as np
```

```python

mask = np.array(Image.open('roku4.png'))
wc = WordCloud(stopwords=STOPWORDS,
               mask=mask, background_color="white",min_word_length=4 ,contour_width=3, contour_color='#c4847c',
               max_words=2500, max_font_size=256,
               random_state=42)
wc.generate(pagina.text)
plt.figure( figsize=(25,20) )
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
mask = np.array(Image.open('eva0.jpg'))
wc2 = WordCloud(stopwords=STOPWORDS,
               mask=mask, background_color="white" ,contour_width=3, contour_color='#c4847c',
               max_words=1000, max_font_size=100,min_word_length=4,
               random_state=42)
wc2.generate(pagina.text)
plt.figure( figsize=(15,10) )
plt.imshow(wc2, interpolation="bilinear")
plt.axis('off')
plt.show()
```
